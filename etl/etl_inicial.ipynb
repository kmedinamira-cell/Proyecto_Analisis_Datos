```python
def aplanar_columnas(columnas):
    new_cols = []
    for col in columnas:
        if isinstance(col,tuple):
            col = [str(c).lower().strip().replace(" ", "_") for c in col
                   if c
                   and
                   "unname" not in str(c).lower()]
            new_cols.append(col[-1])
        else:
            new_cols.append(str(col).lower().strip().replace(" ", "_"))
    return new_cols
```
