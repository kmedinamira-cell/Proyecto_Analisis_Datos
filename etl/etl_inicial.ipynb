```python
#Aplanar columnas
def aplanar_columnas(columnas):
    new_cols = []
    for col in columnas:
        if isinstance(col,tuple):
            col = [str(c).lower().strip().replace(" ", "_") for c in col
                   if c
                   and
                   "unname" not in str(c).lower()]
            new_cols.append(col[-1])
        else:
            new_cols.append(str(col).lower().strip().replace(" ", "_"))
    return new_cols
```

```python
data_saldo_captaciones.columns = aplanar_columnas(columnas=data_saldo_captaciones.columns)
data_distribucion_cartera.columns = aplanar_columnas(columnas=data_distribucion_cartera.columns)
```

```python
#Limpieza de tildes
def limpiar_texto(texto):
    if isinstance(texto, str):
        texto = unicodedata.normalize('NFKD', texto)
        texto = ''.join([c for c in texto if not unicodedata.combining(c)])
    return texto
```

```python
#Llamar función de limpieza
data_saldo_captaciones.columns = data_saldo_captaciones.columns.map(limpiar_texto)
data_distribucion_cartera.columns = data_distribucion_cartera.columns.map(limpiar_texto)
```

```python
municipios = data_saldo_captaciones[["codigo_del_departamento",
                                     "codigo_del_municipio",
                                     "nombre_del_municipio"]]
municipios_unique = (
    municipios
    .drop_duplicates()
    .reset_index(drop=True))
municipios_unique
```

```python
d_departamentos = municipios_unique[
    (municipios_unique["nombre_del_municipio"].str.startswith("TOTAL",na=False)) |
    (municipios_unique["nombre_del_municipio"].str.startswith("BOGOTA",na=False))].copy()

#Limpiar nombres
d_departamentos["nombre_del_municipio"] = (
    d_departamentos["nombre_del_municipio"]
    .str.replace("TOTAL ", "", regex=False)
    .str.lower() # Convert to lowercase for consistent replacement
    .str.replace("nari#o", "Nariño", regex=False) # Replace lowercase 'nari#o'
    .str.title()) # Apply title casing afterwards

#Renombrar columna
d_departamentos = d_departamentos.rename(
    columns={
    "nombre_del_municipio": "nombre_del_departamento"})

# Diccionario departamento -> región
regiones = {
    "Bogota": "Andina",
    "Antioquia": "Andina",
    "Magdalena": "Caribe",
    "Nariño": "Pacífica",
    "Choco": "Pacífica",
    "Cesar": "Caribe",
    "Santander": "Andina",
    "Cordoba": "Caribe",
    "Nacional": "Nacional",
    "Cauca": "Pacífica",
    "Valle Del Cauca": "Pacífica",
    "Caldas": "Andina",
    "Sucre": "Caribe",
    "N. Santander": "Andina",
    "Arauca": "Orinoquía",
    "Caqueta": "Amazonía",
    "Atlantico": "Caribe",
    "Guajira": "Caribe",
    "Boyaca": "Andina",
    "Risaralda": "Andina",
    "Bolivar": "Caribe",
    "Casanare": "Orinoquía",
    "Cundinamarca": "Andina",
    "Quindio": "Andina",
    "Tolima": "Andina",
    "Meta": "Orinoquía",
    "Huila": "Andina",
    "S. Andres Y Prov.": "Insular",
    "Vaupes": "Amazonía",
    "Amazonas": "Amazonía",
    "Guaviare": "Amazonía",
    "Vichada": "Orinoquía",
    "Putumayo": "Amazonía",
    "Guainia": "Amazonía"
}

# Agregar columna región
d_departamentos["region"] = (
    d_departamentos["nombre_del_departamento"]
    .map(regiones)
)

d_departamentos
```

```python
#Diccionario Departamentos y Regiones
dic_departamentos = d_departamentos.set_index(
    'codigo_del_departamento'
    )['nombre_del_departamento']

dic_regiones = d_departamentos.set_index(
    'codigo_del_departamento'
    )['region']
```

```python
#Mapear el nombre del departamento
import pandas as pd

#Mapear el nombre del departamento
data_saldo_captaciones["nombre_del_departamento"] = (
    data_saldo_captaciones["codigo_del_departamento"]
    .map(dic_departamentos)
)

#Mover al lado de la columna codigo_del_departamento
columna = data_saldo_captaciones.pop("nombre_del_departamento")

posicion = (
    data_saldo_captaciones.columns
    .get_loc("codigo_del_departamento") + 1)

data_saldo_captaciones.insert(
    posicion,
    "nombre_del_departamento",
    columna)

# Mapear region
data_saldo_captaciones["region"] = (
    data_saldo_captaciones["codigo_del_departamento"]
    .map(dic_regiones)
)

#Mover al lado de la columna codigo_del_departamento
columna = data_saldo_captaciones.pop("region")

posicion = (
    data_saldo_captaciones.columns
    .get_loc("nombre_del_departamento") + 1)

data_saldo_captaciones.insert(
    posicion,
    "region",
    columna)
```

```python
#Mapear Nombre de las Entidades
nombreentidad = data_distribucion_cartera[["tipo_entidad",
                                     "codigo_entidad",
                                     "nombreentidad"]]

d_entidades = (
    nombreentidad
    .drop_duplicates()
    .reset_index(drop=True))
d_entidades
```

```python
#Diccionario Nombre de Entidades
dic_entidades = d_entidades.set_index(
    "codigo_entidad"
    )["nombreentidad"]
```

```python
#Mapear el nombre de la entidad
data_saldo_captaciones["nombreentidad"] = (
    data_saldo_captaciones["codigo_de_entidad"]
    .map(dic_entidades.to_dict()))

#Mover al lado de la columna codigo_de_entidad
columna_entidad = data_saldo_captaciones.pop("nombreentidad")

posicion_entidad = (
    data_saldo_captaciones.columns
    .get_loc("codigo_de_entidad") + 1)

data_saldo_captaciones.insert(
    posicion_entidad,
    "nombreentidad",
    columna_entidad)
```

```python
 #Filtrar filas donde nombre entidad NO sea NaN
data_saldo_captaciones = (
    data_saldo_captaciones[
        data_saldo_captaciones["nombreentidad"].notna()
    ].reset_index(drop=True))

data_saldo_captaciones
```

```python
data_saldo_captaciones.rename(columns={"nombreentidad": "nombre_entidad"}, inplace=True)
data_distribucion_cartera.rename(columns={"nombreentidad": "nombre_entidad"}, inplace=True)
```
